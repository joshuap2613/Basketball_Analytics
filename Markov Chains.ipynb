```python
#quick script to check something out
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
#only 1 dimensional markov chain
table = pd.read_fwf("Old-Betting/1993.txt",sep="   ",header = None)
```

```python
table
```

```python
arr = np.array(table[4])
```

```python
arr
```

```python
#what follows an O
O = 0
U = 0
totalO = 0
totalU = 0
hist = [0]
for i in range(len(arr)-1):
    if 'O' in str(arr[i]):
        hist.append(hist[-1]+1)
        if 'O' in str(arr[i+1]):
            O +=1
            totalO +=1
        elif 'U' in str(arr[i+1]):
            totalO+=1
    if 'U' in str(arr[i]):
        hist.append(hist[-1]-1)
        if 'U' in str(arr[i+1]):
            U +=1
            totalU +=1
        elif 'O' in str(arr[i+1]):
            totalU +=1
print(O/totalO)
print(U/totalU)
plt.plot(hist)
```
